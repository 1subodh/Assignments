#### Finding outliers

```python
outliers = []
def detect_outlier(data_1):
    outliers = []
    threshold = 3
    mean_1 = np.mean(data_1)
    std_1 = np.std(data_1)   
    for y in data_1:
        z_score = (y - mean_1)/std_1
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers        
```

```python
outlier_datapoints = []
for n in range(0,9):
    outlier_datapoints.extend(detect_outlier(data1[n]))
print((outlier_datapoints,len(outlier_datapoints)))
```

#### Writing ouliers in files

```python
pd.DataFrame(outlier_datapoints).to_csv("./outlier.csv")
```

#### Remove outliers

```python
for n in range(0,9):
    outliers = []
    threshold = 3
    mean_1 = np.mean(df[n])
    std_1 = np.std(df[n])   
    for y in df[n]:
        z_score = (y - mean_1)/std_1
        if np.abs(z_score) > threshold:
            df = df[df[n] != y]
df.shape
```

#### MEAN and MEDIAN

```python
print("Mean:",[non_outlier.mean()[x] for x in non_outlier.columns])
print("Median:",[non_outlier.median()[x] for x in non_outlier.columns])
```

## Mean Absolute Percentage Error (MAPE) - MEAN and MEDIAN

```python
print("Col : MAPE")
for n in non_outlier.columns:
    mape = abs(sum((non_outlier[n] - non_outlier.describe()[n]['mean']) / non_outlier[n] * 100))
    print(n," : ",mape)
```

```python
print("Col : MAPE")
for n in non_outlier.columns:
    mape = abs(sum(non_outlier[n] - non_outlier.median()[n] / non_outlier[n] * 100))
    print(n," : ",mape)
```

## Root MEAN squared error - MEAN and MEDIAN

```python
print("Col : RMSE")
for n in non_outlier.columns:
    rmse = stats.math.sqrt((sum((non_outlier.mean()[n] - non_outlier[n])**2) / len(non_outlier[n]))) 
    print(n," : ",rmse)
```

```python
for n in non_outlier.columns:
    rmse = stats.math.sqrt((sum((non_outlier.median()[n] - non_outlier[n])**2) / len(non_outlier[n]))) 
    print(n," : ",rmse)
```
