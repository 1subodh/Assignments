#### Importing required packages

```python
import gpxpy
import folium
```

####  Opening gpx file

```python
gpxfile = open('./2912087.gpx','r')
```

#### Parseing gpx file

```python
gpx = gpxpy.parse(gpxfile)
```

#### Getting details of file - tracks and segments

```python
track   = gpx.tracks[0]
segment = track.segments[0]
```

#### Points as list of tuples

```python
points  = []
for point in segment.points:
    points.append(tuple([point.latitude, point.longitude]))
```

#### Showing details of WAgpx file

```python
details = {'Tracks' : len(gpx.tracks),'Segments' : len(track.segments),'Points' : len(segment.points)}
details.items()
```

#### Calculating avarage so that initial zoom view can be set

```python
avglat = sum(p[0] for p in points)/len(points)
avglon = sum(p[1] for p in points)/len(points)
```

```python
WAmap = folium.Map(location=[avglat, avglon], zoom_start=15)
```

#### Implementing different folium markers - Arrows and Circles

```python
for each in points:  
    folium.CircleMarker(each, radius = 5, color='red').add_to(WAmap)
    folium.RegularPolygonMarker(location=each, 
                fill_color='black', number_of_sides=1, 
                radius=8).add_to(WAmap)
```

```python
folium.PolyLine(points, color="blue", weight = 5, opacity=20).add_to(WAmap)
#WAmap.save('./WAmap.html')
WAmap
```
